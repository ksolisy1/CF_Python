## Q6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: if the mean prices of products purchased by a user is ower than 10, then flag them as a 'low spender,' if the mean prices of products by a user is higher or equal to 10 then flag them as a 'high spender,'

```python
# creating new column and aggregating data
df_loyal['mean_price'] = df_loyal.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
# create spender_flag based on 'mean price'
df_loyal.loc[df_loyal['mean_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
df_loyal.loc[df_loyal['mean_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# checking output
df_loyal.head(15)
```

```python
# checking value counts
df_loyal['spender_flag'].value_counts(dropna = False)
```

## Q7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# inspecting median of 'days_since_prior_order'
df_loyal.groupby('customer_id').days_since_prior_order.median()
```

```python
# aggregating data with transform and creating 'customer_frequency'
df_loyal['customer_frequency'] = df_loyal.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
# checking output
df_loyal.head(15)
```

```python
# deriving columns with loc() to create frequncy flag
df_loyal.loc[df_loyal['customer_frequency'] >20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
df_loyal.loc[(df_loyal['customer_frequency'] <=20) & (df_loyal['customer_frequency'] > 10), 'frequency_flag'] = 'Regular Customer'
```

```python
df_loyal.loc[df_loyal['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# checking output
df_loyal.head(10)
```

```python
# dropping typo column
df_loyal = df_loyal.drop(columns = ['frequency_flat'])
```

```python
# checking columns
df_loyal.columns
```

```python
# checking flag counts
df_loyal['frequency_flag'].value_counts(dropna = False)
```

```python
# checking missing values
df_loyal.isnull().sum()
```

```python
# creating new flag for missing values - we know that customers with no previous orders are new
df_loyal.loc[df_loyal['customer_frequency'].isnull(), 'frequency_flag'] = 'New customer'
```

```python
# rechecking flag counts
df_loyal['frequency_flag'].value_counts(dropna = False)
```
