## 04 Grouping Data with pandas

```python
# grouping data by 'product_name'
df.groupby('product_name')
```

## 05 Aggregating Data with agg()

```python
# performing single aggregation
df.groupby('department_id').agg({'order_number': ['mean']})
```

## 06 Performing Multiple Aggregations

```python
# performing multiple aggregations
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## 07 Aggregating Data with transform ()

```python
# aggregatig data and creating new column on df_busy
df_busy['max_order'] = df_busy.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
# checking output
df_busy.head(15)
```

## 08 Deriving Columns with loc()

```python
# creating new 'loyalty' flag
df_busy.loc[df_busy['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_busy.loc[(df_busy['max_order'] <= 40) & (df_busy['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_busy.loc[df_busy['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_busy['loyalty_flag'].value_counts(dropna = False)
```
