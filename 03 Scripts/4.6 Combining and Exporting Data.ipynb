```python
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

# 03 Cleaning data to prepare for merge

```python
# check output
df_ords_prod_comb.head()
```

```python
df_ords_prod_comb.shape
```

```python
# dropping merge column from df_ords_prod_comb
df_ords_prod_comb = df_ords_prod_comb.drop(columns = ['_merge'])
```

```python
# rechecking output
df_ords_prod_comb.head()
```

```python
df_ords_prod_comb.shape
```

```python
# checking output of df_prods
df_prods.head()
```

```python
# dropping unnamed column from df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
#check output
df_prods.head()
```

```python
df_prods.shape
```

# 04 Merging Data

```python
# Test merge without overwriting
pd.merge(df_ords_prod_comb,df_prods, on = 'product_id')
```

```python
# merging data
df_ords_prods_merged = df_ords_prod_comb.merge(df_prods, on ='product_id', indicator = True)
```

```python
# checking merge
df_ords_prods_merged.head()
```

```python
df_ords_prods_merged.shape
```

```python
df_ords_prods_merged['_merge'].value_counts()
```

# 04 Exporting data in pickle format

```python
# exporting merged data as pickle
df_ords_prods_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```
